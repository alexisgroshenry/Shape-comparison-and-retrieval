```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_file(file_name) :
    with open(file_name) as f :
        file_data = f.readlines()
    hist = np.zeros((len(file_data),100))
    for ind,k in enumerate(file_data) :
        k = k.split("\t")[:-1]
        for nb,i in enumerate(k) :
            hist[ind,nb] = float(i)
    x = np.linspace(0,100,100)
    for k in range(hist.shape[0]) :
        y = np.array(hist[k,:])
        plt.plot(x,y)
        plt.show()
    return hist
```

```python
hist_octo = read_file("octogon_hist.txt")
```
